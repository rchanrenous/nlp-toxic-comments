```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
x_train = np.load('/content/drive/MyDrive/NLP/bag_of_word/x_train_threshold_100.npy')
y_train = np.load('/content/drive/MyDrive/NLP/bag_of_word/y_train_threshold_100.npy')
```

```python
print('x_train : {}'.format(x_train.shape))
print('y_train : {}'.format(y_train.shape))
```

```python
x_test = np.load('/content/drive/MyDrive/NLP/bag_of_word/x_test_threshold_100.npy')
y_test = np.load('/content/drive/MyDrive/NLP/bag_of_word/y_test_threshold_100.npy')
```

```python
# define the model
def buildModel(train_data):

  model = tf.keras.Sequential()
  model.add(tf.keras.layers.Dense(64, input_dim=train_data.shape[1], activation='relu'))
  model.add(tf.keras.layers.BatchNormalization())
  model.add(tf.keras.layers.Dense(128, activation='relu'))
  model.add(tf.keras.layers.BatchNormalization())
  model.add(tf.keras.layers.Dropout(0.2))
  model.add(tf.keras.layers.Dense(128, activation='relu'))
  model.add(tf.keras.layers.BatchNormalization())
  model.add(tf.keras.layers.Dropout(0.2))
  model.add(tf.keras.layers.Dense(6, activation='sigmoid'))

  return model

model = buildModel(x_train)
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=tf.keras.metrics.AUC())
```

```python
history = model.fit(x_train,y_train, batch_size = 16 ,epochs = 3,validation_split=0.2)
```

```python
print("Evaluate on test data")
results = model.evaluate(x_test, y_test, batch_size=64)
print("test loss, test acc:", results)
```
